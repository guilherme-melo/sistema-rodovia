```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
barbeiro0_5lanes = pd.read_csv('barbeiro0_5lanes.csv')[:-1]
barbeiro_5_3lanes = pd.read_csv('barbeiro5_3lanes.csv')
barbeiro_5_5_lanes = pd.read_csv('barbeiro5_5_lanes.csv')
barbeiro5_lanes5 = pd.read_csv('barbeiro5_lanes5.csv')
barbeiro0_5lanes_50 = pd.read_csv('barbeiro0_5lanes-50.csv')
```

```python
barbeiro0_5lanes
```

```python
barbeiro0_5lanesGrouped = barbeiro0_5lanes.groupby(' Tempo')['N_Rodovias'].mean().reset_index(name = 'tempo_medio')
barbeiro_5_3lanesGrouped = barbeiro_5_3lanes.groupby(' Tempo')['N_Rodovias'].mean().reset_index(name = 'tempo_medio')
barbeiro_5_5_lanesGrouped = barbeiro_5_5_lanes.groupby(' Tempo')['N_Rodovias'].mean().reset_index(name = 'tempo_medio')
barbeiro5_lanes5Grouped = barbeiro5_lanes5.groupby(' Tempo')['N_Rodovias'].mean().reset_index(name = 'tempo_medio')
barbeiro0_5lanes_50 = barbeiro0_5lanes_50.groupby(' Tempo')['N_Rodovias'].mean().reset_index(name = 'tempo_medio')
```

```python
sns.lineplot(x=' Tempo', y='tempo_medio', data=barbeiro0_5lanesGrouped[:-1])
plt.xlabel('Número de rodovias')
plt.ylabel('Tempo médio')
plt.title('Tempo médio 10 pistas e capacidade do barbeiro 0')
plt.savefig('images/barbeiro0_5lanes.png')
```

```python
sns.lineplot(x=' Tempo', y='tempo_medio', data=barbeiro_5_3lanesGrouped)
plt.xlabel('Número de rodovias')
plt.ylabel('Tempo médio')
plt.title('Tempo médio 6 pistas e capacidade do barbeiro 5')
plt.savefig('images/barbeiro_5_3lanes.png')
```

```python
sns.lineplot(x=' Tempo', y='tempo_medio', data=barbeiro_5_5_lanesGrouped)
plt.xlabel('Número de rodovias')
plt.ylabel('Tempo médio')
plt.title('Tempo médio 10 pistas e capacidade do barbeiro 5')
plt.savefig('images/barbeiro_5_5_lanes.png')
```

```python
barbeiro5_lanes5Grouped = barbeiro5_lanes5Grouped[:-1]
sns.lineplot(x=' Tempo', y='tempo_medio', data=barbeiro5_lanes5Grouped)
plt.xlabel('Número de rodovias')
plt.ylabel('Tempo médio')
plt.title('Tempo médio 10 pistas e capacidade do barbeiro 5')
plt.savefig('images/barbeiro5_lanes5.png')
```

```python
sns.lineplot(x=' Tempo', y='tempo_medio', data=barbeiro0_5lanes_50)
plt.xlabel('Número de rodovias')
plt.ylabel('Tempo médio')
plt.title('Tempo médio 10 pistas e capacidade do barbeiro 0')
plt.savefig('images/barbeiro0_lanes5-50.png')
```
